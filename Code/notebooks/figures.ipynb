```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["font.size"] = 16
```

```python
# y = f(x)
x = np.array([0, 1, 2])
ymf = np.array([1.401, 1.020, 0.919])
ylda = np.array([1.395, 1.019, 0.874])
ydeepconn = np.array([1.675, 1.066, 0.854])


# figure
fig = plt.figure()
fig.subplots_adjust(left=0.15)
ax = fig.add_subplot(1, 1, 1)

# plot
ax.plot(x, ymf, marker='o', linestyle='-', color='b', label='MF')
ax.plot(x, ylda, marker='o', linestyle='-', color='r', label='LDA')
ax.plot(x, ydeepconn, marker='o', linestyle='--', color='g', label='DeepCoNN')

# x axis
plt.xlim([-0.2, 2.2])
ax.set_xticks([0, 0.5, 1, 1.5, 2])
ax.set_xticklabels(["Am_1", "", "Am_5", "", "Am_10"])
#ax.set_xlabel('x')

# y axis
plt.ylim([0.8, 1.70])
ax.set_yticks([1.0, 1.25, 1.5])
ax.set_ylabel('MSE')

# legend and title
ax.legend(loc='best')
ax.set_title('Amazon')
plt.grid()
# save as png
plt.savefig('../../../MyThesis/figures/amazon.png', dpi=300.0)

```
